```python
import pandas as pd
import requests
import json

download_url = 'https://lantern.healthit.gov/api/daily/download'

lantern_df = pd.read_csv(download_url)

org_lists = lantern_df['list_source'].unique()

done_lists = []
entries = []
errors=[]
```

```python
i=0
for item in org_lists:
    if item not in done_lists:
        if i==42 or i==82 or i ==212:
            errors.append({'source':item, 'error': 'ConnectionResetError'})
        elif i==127:
            errors.append({'source':item, 'error': 'CMS Blocked Site'})
        else:
            try:
                resp = requests.get(item, verify=False).json()
                resp_entry = resp['entry']
                if type(resp['entry']) != list:
                    resp_entry = list(resp_entry)
                new_entries = [{**entry['resource'], 'source': list} for entry in resp_entry]
                entries+=new_entries
            except Exception as e:
                errors.append({'source':item, 'error': e})
        done_lists.append(item)
    i+=1
    print(i)
```
